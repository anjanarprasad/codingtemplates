Load the Libraries

EDA is classified into four types
1)univariate non-graphical
2)Univariate graphical
3)Multivariate non-graphical
4)Multivariate graphical

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

Reading the Data

```python
df = pd.read_csv("yourdata.csv")
print("shape :", df.shape)
df.head()
```

Checking for NULL and Duplicate Values

Univariate Non Graphical

```python
print("datatypes\n",df.dtypes)
print("\n missing values : \n",df.isnull.sum())
print("\n duplicate values :\n",df.duplicated.sum())
print("\n summary statistics : \n", df.describe(include = "all"))
```

PLOTS

Univariate Graphical

```python
df.hist(bins=30,figsize=(15,10))
plt.tight_layout()
plt.show()

numeric_cols = df.select_dtypes(include = np.number).columns
for col in numeric_cols:
  sns.boxplot(x=df[col])
  plt.title(col)
  plt.show()
```

Multivariate Non-Graphical

```python
df.groupby("category column")["numeric column"].mean()
df.groupby("category_column")["numeric column"].agg(["mean", "median","min", "max"] )
pd.crosstab(df['category1'],df['category2'])
df.corr(numeric_only = true)
df.pivot_table(value='value_column',index='group_column', columns='category_column')
```

Multivariate Graphical

```python
sns.scatterplot(x='numeric_col1',y='numeric_col2',hue='category_col', data=df)
plt.title("scatterplot")
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("correlation heatmap")
sns.pairplot(df[['numeric_col1','numeric_col2', 'numeric_col3']]), hue='target_col')
plt.suptitle("Pairplot",y=1.02)
sns.boxplot(x="category_col", y="numeric_col", data=df)
plt.title("boxplot")
sns.violinplot(x="category_col",y="numeric_col", data=df)
plt.title("violinplot")
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot= true, fmt=".2f", cmap="coolwarm")
plt.title("correlation heatmap")
plt.show()
```
